```python
#Creating a Function (With Dataset as Parameter) for the Above Logics Applied with Return Statements of Created List(s) - quan, qual

def quanQual(dataset):
    
    quan=[]
    qual=[]
    
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
"""
Dataset (With Outliers)
To Automate the "Mean","Median","Mode" and "Level of Percentile"("Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%"),
"IQR","1.5_rule","LesserOutlier","GreaterOutlier","MinValue","MaxValue" to be filled in Table automatically
"""

descriptive = pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5_rule",
                                  "LesserOutlier","GreaterOutlier","MinValue","MaxValue"],columns=quan)

for columnName in quan:
    descriptive[columnName]["Mean"] = dataset[columnName].mean()
    descriptive[columnName]["Median"] = dataset[columnName].median()
    descriptive[columnName]["Mode"] = dataset[columnName].mode()[0]
    descriptive[columnName]["Q1:25%"] = dataset.describe()[columnName]["25%"]
    descriptive[columnName]["Q2:50%"] = dataset.describe()[columnName]["50%"]
    descriptive[columnName]["Q3:75%"] = dataset.describe()[columnName]["75%"]
    descriptive[columnName]["99%"] = np.percentile(dataset[columnName],99)
    descriptive[columnName]["Q4:100%"] = dataset.describe()[columnName]["max"]
    
    descriptive[columnName]["IQR"] = descriptive[columnName]["Q3:75%"] - descriptive[columnName]["Q1:25%"]
    descriptive[columnName]["1.5_rule"] = 1.5 * descriptive[columnName]["IQR"]
    descriptive[columnName]["LesserOutlier"] = descriptive[columnName]["Q1:25%"] - descriptive[columnName]["1.5_rule"]
    descriptive[columnName]["GreaterOutlier"] = descriptive[columnName]["Q3:75%"] + descriptive[columnName]["1.5_rule"]
    descriptive[columnName]["MinValue"] = dataset[columnName].min()
    descriptive[columnName]["MaxValue"] = dataset[columnName].max()

descriptive
```

```python
"""Creating a Function To Automate the "Mean","Median","Mode" and "Level of Percentile"("Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%"),
"IQR","1.5_rule","LesserOutlier","GreaterOutlier","MinValue","MaxValue", "Skewness", "Kurtosis", "Variance", "StdDeviation" 
to be filled in Table automatically """

def Univariate(dataset,quan):
    descriptive = pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5_rule",
                        "LesserOutlier","GreaterOutlier","MinValue","MaxValue","Skewness","Kurtosis",
                                      "Variance","StdDeviation"],columns=quan)

    for columnName in quan:
            descriptive[columnName]["Mean"] = dataset[columnName].mean()
            descriptive[columnName]["Median"] = dataset[columnName].median()
            descriptive[columnName]["Mode"] = dataset[columnName].mode()[0]
            descriptive[columnName]["Q1:25%"] = dataset.describe()[columnName]["25%"]
            descriptive[columnName]["Q2:50%"] = dataset.describe()[columnName]["50%"]
            descriptive[columnName]["Q3:75%"] = dataset.describe()[columnName]["75%"]
            descriptive[columnName]["99%"] = np.percentile(dataset[columnName],99)
            descriptive[columnName]["Q4:100%"] = dataset.describe()[columnName]["max"]

            descriptive[columnName]["IQR"] = descriptive[columnName]["Q3:75%"] - descriptive[columnName]["Q1:25%"]
            descriptive[columnName]["1.5_rule"] = 1.5 * descriptive[columnName]["IQR"]
            descriptive[columnName]["LesserOutlier"] = descriptive[columnName]["Q1:25%"] - descriptive[columnName]["1.5_rule"]
            descriptive[columnName]["GreaterOutlier"] = descriptive[columnName]["Q3:75%"] + descriptive[columnName]["1.5_rule"]
            descriptive[columnName]["MinValue"] = dataset[columnName].min()
            descriptive[columnName]["MaxValue"] = dataset[columnName].max()
            
            descriptive[columnName]["Skewness"] = dataset[columnName].skew()
            descriptive[columnName]["Kurtosis"] = dataset[columnName].kurtosis()
            
            descriptive[columnName]["Variance"] = dataset[columnName].var()
            descriptive[columnName]["StdDeviation"] = dataset[columnName].std()
        
    return descriptive
```

```python
"""Creating a Function To Automate the Updatation of Frequency, Relative Frequency and Cumulative Relative Frequency 
by Creating a New Table - FrequencyTable """

def frequencyTable(dataset,columnName):
    frequencyTable = pd.DataFrame(columns=["Unique_Values","Frequency","Relative_Frequency","CumSum"])
    frequencyTable["Unique_Values"] = dataset[columnName].value_counts().index
    frequencyTable["Frequency"] = dataset[columnName].value_counts().values
    frequencyTable["Relative_Frequency"] = (frequencyTable["Frequency"]/103)
    frequencyTable["CumSum"] = frequencyTable["Relative_Frequency"].cumsum() #Calling an Existing Function .cumsum()
    return frequencyTable
```
