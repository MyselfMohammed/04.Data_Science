```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Placement_Data_Full_Class.csv')
```

```python
import pandas as pd
import numpy as np
from pandas.api.types import CategoricalDtype
from plotnine import *
from plotnine.data import mpg
%matplotlib inline

(ggplot(dataset)         # defining what data to use
 + aes(x='gender')    # defining what variable to use
 + geom_bar(size=20) # defining the type of plot to use
)
```

```python
(ggplot(dataset)
 + aes(x='gender')
 + geom_bar(size=20)
 + coord_flip()        # flipping the x- and y-axes
 + labs(title='Number of Male and Female', x='Gender', y='count') # customizing labels
)
```

```python

(ggplot(dataset)
 + aes(x='gender',y='ssc_p',fill='ssc_b')
 + geom_boxplot(size=20)
       # flipping the x- and y-axes
 + labs(title='Number of Male and Female', x='Gender', y='count') # customizing labels
)
```

```python
(ggplot(dataset, aes(x='continent', y='lifeExp', fill='continent')) +
        geom_boxplot()
)
```
